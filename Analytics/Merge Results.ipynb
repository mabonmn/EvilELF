```python
import pandas as pd
import math
import numpy as np
import os
```

```python
directory = "/home/andrewkosikowski/Documents/predictions/original_malware_test_results/"
files = os.listdir(directory)
first = True
df = pd.DataFrame()
index = 2;
for file in files:
    if '.csv' in file:
        file_path = os.path.join(directory, file)
        if os.path.isfile(file_path):
            df2 = pd.read_csv(file_path) 
            if first:
                df = df2
                df.columns.values[index] = file
                index = index + 1
                first = False
                continue
            df = pd.concat([df, df2[['Prediction']]], axis=1)
            df
            df.columns.values[index] = file
            index = index + 1
```

```python
total_true_predictions = df.iloc[:,2:].sum(axis=1,numeric_only=True)
num_models = df.shape[1]-2
mispredictions = (df['ground_truth'] * num_models - total_true_predictions).abs()
df.insert(loc = 2,
          column = 'correct_predictions',
          value = num_models - mispredictions)
df.insert(loc = 3,
          column = 'mispredictions',
          value = mispredictions)
```
